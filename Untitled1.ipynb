```python
import pandas as pd
import altair as alt

data = pd.read_csv("sapi_2022.csv")
df = pd.DataFrame(data)

df_enroll = df.iloc[5:48,0:2:1]

df_tsukukoma = df.iloc[5:48,0:3:2]

## school_name=['東京','自由が丘','吉祥寺','横浜','巣鴨','練馬','大井町','王子','用賀','大船','白銀高輪','西船','渋谷','たまプラーザ(宮前平と合算)','お茶の水','中野','南浦和','武蔵小杉','豊洲','柏','千葉','日吉','成城','国立','新浦安','青葉台','仙川','町田','高田馬場','海浜幕張','下高井戸','茗荷谷','永福町','センター南','大宮','東戸塚','若葉台','上大岡','宮前平(閉鎖)','茅ヶ崎','所沢','新越谷','松戸']
enrollment=['校舎名','在籍者数']
successful_candidate=['校舎名','合格者数']
```

```python
df_enroll.columns = enrollment
df_enroll
```

```python
df_tsukukoma.columns = successful_candidate
df_tsukukoma
```

```python
chart = (
    alt.Chart(df_enroll)
    .mark_bar(opacity=0.8, clip=True)
    .encode(
        x='校舎名',
        y=alt.Y('在籍者数:Q', stack=None),
        color='Name:N'
    )
)
chart
```

```python
chart = (
    alt.Chart(df_tsukukoma)
    .mark_line(opacity=0.8, clip=True)
    .encode(
        x='校舎名',
        y=alt.Y('合格者数:Q', stack=None),
        color='Name:N'
    )
)
chart
```

```python
df_tsukukoma2 = df_tsukukoma.drop('校舎名', axis=1)
df_tsukukoma2
```

```python
df_concat = pd.concat([df_enroll,df_tsukukoma2], axis=1)
df_concat
```

```python
df_concat = pd.melt(df_concat, id_vars=['校舎名'])
df_concat
```

```python
df_concat = df_concat.rename(
    columns={'value':'該当者数'}
)
df_concat
```

```python
variable = ['在籍者数','合格者数']
```

```python
chart = (
    alt.Chart(df_concat)
    .mark_bar(opacity=0.8, clip=True)
    .encode(
        x='校舎名',
        y=alt.Y('該当者数:Q', stack=None),
        color=alt.Color('variable', scale=alt.Scale(domain=variable, range=["blue", "red"]))
    )
)
chart
```

```python
def create_graph():
    chart = (
    alt.Chart(df_concat)
    .mark_bar(opacity=0.8, clip=True)
    .encode(
        x='校舎名',
        y=alt.Y('該当者数:Q', stack=None),
        color=alt.Color('variable', scale=alt.Scale(domain=variable, range=["blue", "red"]))
        )
    )
    return chart
```

```python
create_graph()
```
